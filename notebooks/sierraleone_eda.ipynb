```python

import sys
sys.path.append("..")  

from src.data_loader import load_data
from src.data_cleaning import clean_data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Load Data ---
file_path = "../data/sierraleone-bumbuna.csv"
df = load_data(file_path)

print(f"✅ Data loaded: {df.shape[0]} rows × {df.shape[1]} columns")

# Preview
display(df.head())
display(df.info())

# --- 2. Data Cleaning ---
clean_df = clean_data(df)

print(f"✅ Cleaning complete: {clean_df.shape[0]} rows × {clean_df.shape[1]} columns")

# --- 3. Summary Statistics ---
display(clean_df.describe().T)

# --- 4. Missing Values Report ---
missing_report = clean_df.isna().sum().to_frame("missing_count")
missing_report["missing_%"] = (missing_report["missing_count"] / len(clean_df)) * 100
display(missing_report[missing_report["missing_%"] > 0])

# --- 5. Correlation Heatmap ---
# Adjusted version to avoid overlapping labels
plt.figure(figsize=(10, 8))
corr = clean_df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="YlGnBu",
    annot_kws={"size": 8},
    cbar_kws={'shrink': 0.8}
)

plt.title("Correlation Heatmap - Sierraleone", fontsize=14, pad=20)
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.yticks(fontsize=9)
plt.tight_layout()
plt.show()


# --- 6. Save Cleaned Data ---
output_path = "../data/cleaned/sieraleone_clean.csv"
clean_df.to_csv(output_path, index=False)

print(f"💾 Cleaned dataset saved to: {output_path}")
```
